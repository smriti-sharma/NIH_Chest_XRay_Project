```
Labels = {'Pneumothorax', 'Effusion', 'Atelectasis', 'Mass', 'NoFinding',
       'Pleural_Thickening', 'Nodule', 'Fibrosis', 'Edema', 'Infiltration',
       'Cardiomegaly', 'Hernia', 'Pneumonia', 'Emphysema', 'Consolidation'}

indices = []
count = 0
for i in rules.index:
   cons = rules.at[i, "consequence"]           
   cons = cons.replace("{","")
   cons = cons.replace("}","")
   cons = cons.replace(" ","")
   cons = set(cons.split(","))
   if(cons.issubset(Labels)):
     indices.append(i)
     if(len(cons)>1):
       count = count+1
       #print(cons)

print(len(indices))

print(count)
final_rules = rules.iloc[indices]
#print(final_rules)
```
